```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt


def f(x):
    return x[0] ** 3 + x[1] ** 3 + 3. * x[0] * x[1] ** 2 - 15. * x[0] - 15. * x[1]
```

```python
def decorate_axes(x_label=None, y_label=None, fontsize=16):
    plt.xticks(fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    if x_label is not None:
        plt.xlabel(x_label, fontsize=fontsize)
    if y_label is not None:
        plt.ylabel(y_label, fontsize=fontsize)
```

```python
def draw_contours(fct, x_range=(-3., 3.), y_range=(-3., 3.), levels=20):
    step = 0.01
    x = np.arange(x_range[0], x_range[1], step)
    y = np.arange(y_range[0], y_range[1], step)
    xg, yg = np.meshgrid(x, y)
    zg = fct((xg, yg))
    plt.figure(figsize=(10, 10), tight_layout=True)
    cp = plt.contour(xg, yg, zg, levels=levels)
    plt.clabel(cp, inline=True, fontsize=16)
    decorate_axes()
    plt.show()
```

```python
draw_contours(f)
```

### Problem 1.5
**Create functions gradient(fct, x0, delta) and hessian(fct, x0, delta)**

```python
def gradient(fct, x0, delta): 
    n = len(x0)
    final_gradient = []
    for i in range(0, n): 
        upper = list(x0)
        lower = list(x0)
        upper[i] = upper[i] + delta #f(x + delta, y, ...)
        lower[i] = lower[i] - delta #f(x - delta, y, ...)
        
        partial_derivative = ( fct(tuple(upper)) - fct(tuple(lower)) ) / (2 * delta)
        final_gradient.append(partial_derivative)
        
    return np.array(final_gradient)
        
        
```

```python
def hessian(fct, x0, delta): 
    n = len(x0)
    final_hessian = []
    for i in range(0, n): 
        upper = list(x0)
        lower = list(x0)
        upper[i] = upper[i] + delta
        lower[i] = lower[i] - delta
        partial_upper = gradient(fct, upper, delta) #f_x(x + delta, y, ...)
        partial_lower = gradient(fct, lower, delta) #f_x(x - delta, y, ...)
        
        current_hessian = (partial_upper - partial_lower) / (2 * delta)
        final_hessian.append(current_hessian)
    return np.array(final_hessian)

    
```
